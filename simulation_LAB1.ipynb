## THEORY:
Chemical reactions in dynamic equilibrium involve both forward and backward reactions. At a steady state, the rates of these reactions are balanced. Let's take the example reaction Ch1 + Ch2 → Ch3. The reaction rate is influenced by factors like the amounts of Ch1 and Ch2, temperature, pressure, humidity, and the catalyst used.
 Neglecting the effects like temperature ,humidity, etc. we simplify our model with a set of  differential equations that describe the rates of increase for Ch1, Ch2, and Ch3 are as follows:

$$\frac{dC1}{dt} = K2C3 - K1C1C2.......(i)$$<br>
$$\frac{dC2}{dt} = K2C3 - K1C1C2........(ii)$$<br>
$$\frac{dC3}{dt} = 2K1C1C2 - 2K2C3........(iii)$$<br>

Here, K1 and K2 are constants. The simulation involves determining the system's state, i.e., the values of C1, C2, and C3 at different time points. The simulation begins at zero time, with small time increments considered for each step.

```python
!pip install ipywidgets
import matplotlib.pyplot as plt # for static plots
import ipywidgets as widgets  # for interactive plots
from ipywidgets import interact, FloatSlider
```

Initialize amount of reactants and rate of reaction

```python
c1=1  #initial amount
c2=2
c3=3
k1=0.07
k2=0.01
k3=1
```

```python
time=[]
amount_c1=[]
amount_c2=[]
amount_c3=[]
t=0
delt = 0.1
while(t<10):
    dc1=k2*c3-k1*c1*c2
    dc2=k2*c3-k1*c1*c2
    dc3=2*k1*c1*c2-2*k2*c3
    c1+=dc1
    c2+=dc2
    c3+=dc3
    amount_c1.append(c1)
    amount_c2.append(c2)
    amount_c3.append(c3)
    time.append(t)
    t+=delt
```

## Result:
The amount of reactants is plotted with respect to time. The amount grows or reduces smoothly depending upon the values of initial amounts and constants(k1,k2,k3).We can observe that the trend of C1 and C2 is the same and C3 is exact opposite.for example when C1 increases C2 also increases and C3 decreases.Similarly if C3 is increasing C1 and C2 are decreasing.

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, figsize=(8, 12)) # plot the values
ax1.plot(time, amount_c1, label='Amount C1', color='blue')
ax1.set_title('Amount C1')
ax1.legend()
ax2.plot(time, amount_c2, label='Amount C2', color='green')
ax2.set_title('Amount C2')
ax2.legend()
ax3.plot(time, amount_c3, label='Amount C3', color='red')
ax3.set_title('Amount C3')
ax3.legend()
plt.xlabel('Time')
plt.show()
```

##**Making the plot more interactive**

```python
def plot_reaction(c1=1, c2=2, c3=3, k1=0.07, k2=0.01, k3=1):
    time=[]
    amount_c1=[]
    amount_c2=[]
    amount_c3=[]
    t=0
    delt = 0.1
    while(t<10):
        dc1=k2*c3-k1*c1*c2
        dc2=k2*c3-k1*c1*c2
        dc3=2*k1*c1*c2-2*k2*c3
        c1+=dc1
        c2+=dc2
        c3+=dc3
        amount_c1.append(c1)
        amount_c2.append(c2)
        amount_c3.append(c3)
        time.append(t)
        t+=delt

    plt.figure(figsize=(10,6))
    plt.plot(time, amount_c1, label='c1',color='red',marker="*")
    plt.plot(time, amount_c2, label='c2',color='blue',marker="+")
    plt.plot(time, amount_c3, label='c3',color='green',marker="o")
    plt.xlabel('Time')
    plt.ylabel('Amount')
    plt.title('Amount of reactants during a chemical reaction')
    plt.legend()
    plt.grid(True)
    plt.show()

w = widgets.interactive(plot_reaction,
         c1=FloatSlider(min=0, max=5, step=0.1, value=1),
         c2=FloatSlider(min=0, max=5, step=0.1, value=2),
         c3=FloatSlider(min=0, max=5, step=0.1, value=3),
         k1=FloatSlider(min=0, max=1, step=0.01, value=0.07),
         k2=FloatSlider(min=0, max=1, step=0.01, value=0.01),
         k3=FloatSlider(min=0, max=1, step=0.1, value=1))

display(widgets.HBox([w.children[-1],widgets.VBox([w.children[i] for i in range(len(w.children)-1)])]))
```
