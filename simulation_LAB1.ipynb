```python
!pip install ipywidgets
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
```

Initialize amount of reactants and rate of reaction

```python
c1=1  #initial amount
c2=2
c3=3
k1=0.07 #rate of reaction
k2=0.01
k3=1
```

Calculate amount of reactants at different time period

```python
time=[]
amount_c1=[]
amount_c2=[]
amount_c3=[]
t=0
delt = 0.1
while(t<10):
    dc1=k2*c3-k1*c1*c2
    dc2=k2*c3-k1*c1*c2
    dc3=2*k1*c1*c2-2*k2*c3
    c1+=dc1
    c2+=dc2
    c3+=dc3
    amount_c1.append(c1)
    amount_c2.append(c2)
    amount_c3.append(c3)
    time.append(t)
    t+=delt
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, figsize=(8, 12)) # plot the values
ax1.plot(time, amount_c1, label='Amount C1', color='blue')
ax1.set_title('Amount C1')
ax1.legend()
ax2.plot(time, amount_c2, label='Amount C2', color='green')
ax2.set_title('Amount C2')
ax2.legend()
ax3.plot(time, amount_c3, label='Amount C3', color='red')
ax3.set_title('Amount C3')
ax3.legend()
plt.xlabel('Time')
plt.show()
```

##**Making the plot more interactive**

```python
import ipywidgets as widgets
def plot_reaction(c1=1, c2=2, c3=3, k1=0.07, k2=0.01, k3=1):
    time=[]
    amount_c1=[]
    amount_c2=[]
    amount_c3=[]
    t=0
    delt = 0.1
    while(t<10):
        dc1=k2*c3-k1*c1*c2
        dc2=k2*c3-k1*c1*c2
        dc3=2*k1*c1*c2-2*k2*c3
        c1+=dc1
        c2+=dc2
        c3+=dc3
        amount_c1.append(c1)
        amount_c2.append(c2)
        amount_c3.append(c3)
        time.append(t)
        t+=delt

    plt.figure(figsize=(10,6))
    plt.plot(time, amount_c1, label='c1',color='red',marker="*")
    plt.plot(time, amount_c2, label='c2',color='blue',marker="+")
    plt.plot(time, amount_c3, label='c3',color='green',marker="o")
    plt.xlabel('Time')
    plt.ylabel('Amount')
    plt.title('Amount of reactants during a chemical reaction')
    plt.legend()
    plt.grid(True)
    plt.show()

w = widgets.interactive(plot_reaction,
         c1=FloatSlider(min=0, max=5, step=0.1, value=1),
         c2=FloatSlider(min=0, max=5, step=0.1, value=2),
         c3=FloatSlider(min=0, max=5, step=0.1, value=3),
         k1=FloatSlider(min=0, max=1, step=0.01, value=0.07),
         k2=FloatSlider(min=0, max=1, step=0.01, value=0.01),
         k3=FloatSlider(min=0, max=1, step=0.1, value=1))

display(widgets.HBox([w.children[-1],widgets.VBox([w.children[i] for i in range(len(w.children)-1)])]))
```
